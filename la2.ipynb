```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import zscore
```

```python
sData = pd.read_csv('/content/Sales.csv')
sData
```

```python
sData['Revenue'] = sData['Price'] * sData['QuantitySold']
```

```python

sData['Price_zscore'] = zscore(sData['Price'])

sData = sData[(sData['Price_zscore'] > -3) & (sData['Price_zscore'] < 3)]

sData.drop(columns=['Price_zscore'], inplace=True)
```

```python
scaler = MinMaxScaler()

sData[['Price', 'Revenue']] = scaler.fit_transform(sData[['Price', 'Revenue']])
```
