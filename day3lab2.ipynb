```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
mnist_df=pd.read_csv('/content/mnist.csv')
mnist_df.head()
```

```python
features=mnist_df.drop('label',axis=1)
labels=mnist_df['label']
```

```python
#  normalize pixel values to be in range (0,1)
features = features / 255.0
```

```python
features=np.array(features)
labels=np.array(labels)
```

```python
features=features.reshape(-1,28,28)
```

```python
labels=labels.reshape(-1)
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# prompt: train the model

history=model.fit(features, labels, epochs=5)
```

```python
predictions=model.predict(features)
```
